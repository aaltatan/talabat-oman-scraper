```python
import pandas as pd
import json
```

```python
def get_data():
    with open('items_with_categories.jsonl', 'r', encoding='utf-8') as file:
        for _ in range(10_000):
            line = file.readline()
            row = json.loads(line)
            yield row
```

```python
df = pd.json_normalize(get_data())
```

```python
pd.set_option('display.max_columns', 100)
```

```python
df = df.drop(['attributes', 'id'], axis='columns')
```

```python
df_images = df['images']
```

```python
df_images = df_images.apply(lambda x: ",".join(x)).astype(str).str.split(',', expand=True)
```

```python
df_images.columns = [f'image-{idx}' for idx, image in enumerate(df_images.columns, start=1)]
```

```python
df = pd.concat([df, df_images], axis='columns')
```

```python
df = df.drop(['images', 'subcategory.subCategories'], axis='columns')
```
